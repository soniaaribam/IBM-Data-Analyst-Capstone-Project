### Task 1: Trends in Compensation Over Age Groups


##### 1. Line Chart of Median `ConvertedCompYearly` by Age Group


- Track how the median yearly compensation (ConvertedCompYearly) changes across different age groups.

- Use a line chart to visualize these trends.


```python
## Write your code here
df_filtered = df[['Age', 'ConvertedCompYearly']].dropna()
comp_by_age = df_filtered.groupby('Age')['ConvertedCompYearly'].median()
plt.figure(figsize=(10, 6))
plt.plot(comp_by_age.index, comp_by_age.values, marker='o', linestyle='-')

# Labels and title
plt.xlabel('Age Group')
plt.ylabel('Median Yearly Compensation')
plt.title('Trends in Compensation Over Age Groups')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.grid(True)
```

##### 2. Line Chart of Median `ConvertedCompYearly` for Ages 25 to 45


For a closer look, plot a line chart focusing on the median compensation for respondents between ages 25 and 45.


```python
## Write your code here
age_filter = ['25-34 years old', '35-44 years old']
df_filtered = df[df['Age'].isin(age_filter)][['Age', 'ConvertedCompYearly']].dropna()

# Group by Age and calculate median compensation
comp_by_age = df_filtered.groupby('Age')['ConvertedCompYearly'].median()

# Plotting the line chart
plt.figure(figsize=(8, 5))
plt.plot(comp_by_age.index, comp_by_age.values, marker='o', linestyle='-', color='b')

# Labels and title
plt.xlabel('Age Group')
plt.ylabel('Median Yearly Compensation')
plt.title('Median Compensation for Ages 25 to 45')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.grid(True)

# Show plot
plt.show()
```

### Task 2: Trends in Job Satisfaction by Experience Level



##### 1. Line Chart of Job Satisfaction (`JobSatPoints_6`) by Experience Level



- Use a column that approximates experience level to analyze how job satisfaction changes with experience.

- If needed, substitute an available experience-related column for `Experience`.


```python
## Write your code here
df['YearsCodePro'] = pd.to_numeric(df['YearsCodePro'], errors='coerce')

# Drop NaN values in relevant columns
df_filtered = df[['YearsCodePro', 'JobSatPoints_6']].dropna()

# Group by experience level and calculate median job satisfaction
job_sat_by_experience = df_filtered.groupby('YearsCodePro')['JobSatPoints_6'].median()

# Plotting the line chart
plt.figure(figsize=(10, 6))
plt.plot(job_sat_by_experience.index, job_sat_by_experience.values, marker='o', linestyle='-', color='g')

# Labels and title
plt.xlabel('Years of Professional Experience')
plt.ylabel('Median Job Satisfaction (JobSatPoints_6)')
plt.title('Trends in Job Satisfaction by Experience Level')
plt.grid(True)

# Show plot
plt.show()
```

### Task 3: Trends in Job Satisfaction and Compensation by Experience


##### 1.Line Chart of Median ConvertedCompYearly Over Experience Level

- This line chart will track how median compensation (`ConvertedCompYearly`) changes with increasing experience.

- Use a column such as `WorkExp` or another relevant experience-related column.


```python
## Write your code here
df['YearsCodePro'] = pd.to_numeric(df['YearsCodePro'], errors='coerce')

# Drop NaN values for relevant columns
df_filtered = df[['YearsCodePro', 'ConvertedCompYearly']].dropna()

# Group by experience level and calculate median compensation
comp_by_experience = df_filtered.groupby('YearsCodePro')['ConvertedCompYearly'].median()

# Plotting the line chart
plt.figure(figsize=(10, 6))
plt.plot(comp_by_experience.index, comp_by_experience.values, marker='o', linestyle='-', color='b')

# Labels and title
plt.xlabel('Years of Professional Experience')
plt.ylabel('Median Yearly Compensation')
plt.title('Trends in Compensation Over Experience Level')
plt.grid(True)

# Show plot
plt.show()
```
